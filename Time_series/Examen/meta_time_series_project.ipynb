```python
df = pd.read_csv('social_networks.csv')
buttons =[]
df = df.drop(columns=["isPartial"])
fig = px.line(df, x="date", y=df.columns, title="Meta Social Media Trends")


buttons.append(dict(method='update',
                        label="All",
                        args = [{'visible': [True for r in df.drop(columns=["date"])]}]
                        ))
for rs in df.drop(columns=["date"]):
    buttons.append(dict(method='update',
                        label=rs,
                        args = [{'visible': [rs == r for r in df.drop(columns=["date"])]}]
                        ))
# Add dropdown menus to the figure
fig.update_layout(showlegend=False, updatemenus=[{"buttons": buttons, "direction": "down", "showactive": True, "x": 0.5, "y": 1.15}])
fig.show()
```
