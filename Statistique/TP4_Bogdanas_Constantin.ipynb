En utilisant les commandes M = runif(n∧2, 0, 1) puis N = matrix(M, n, n), on peut générer une matrice carrée

3.  Faîtes la même chose en considérant maintenant de matrices symétriques (trouver une astuce pour en simuler facilement...). Résultat? Est-ce la même vitesse? Utiliser alors l'inversion par la commande chol2inv. Résultat? Et qu'en est-il du calcul du déterminant ?


```r
 #  1. Faîtes un programme qui inverse la matrice N pour diérentes valeurs de n et mémoriser le temps de calcul à chaque fois. 
 # J'ai fais que pour une matrice 3x3 n = 3 
invert_matrix3x3 <- function(matA) {
  if (ncol(matA)!=3 | nrow(matA)!=3) stop("The matrix is not a 3x3 matrix")
  b <- matrix(rep(0,9), ncol=3)
  det1a <- (matA[1,3]*matA[2,2]*matA[3,1] - matA[1,2]*matA[2,3]*matA[3,1] - matA[1,3]*matA[2,1]*matA[3,2] + matA[1,1]*matA[2,3]*matA[3,2] + matA[1,2]*matA[2,1]*matA[3,3] - matA[1,1]*matA[2,2]*matA[3,3])
  det2a <- -det1a
  if (det1a == 0) stop("The matrix is not invertible")
  numb11 <- (matA[2,3]*matA[3,2] - matA[2,2]*matA[3,3])
  numb21 <- (matA[2,1]*matA[3,3] - matA[2,3]*matA[3,1])
  numb31 <- (matA[2,1]*matA[3,2] - matA[2,2]*matA[3,1])
  b[1,1] <- numb11 / det1a
  b[2,1] <- numb21 / det1a
  b[3,1] <- numb31 / det2a
  numb12 <- (matA[1,3]*matA[3,2] - matA[1,2]*matA[3,3])
  numb22 <- (matA[1,3]*matA[3,1] - matA[1,1]*matA[3,3])
  numb32 <- (matA[1,1]*matA[3,2] - matA[1,2]*matA[3,1])
  b[1,2] <- numb12 / det2a
  b[2,2] <- numb22 / det1a
  b[3,2] <- numb32 / det1a
  numb13 <- (matA[1,3]*matA[2,2] - matA[1,2]*matA[2,3])
  numb23 <- (matA[1,1]*matA[2,3] - matA[1,3]*matA[2,1])
  numb33 <- (matA[1,1]*matA[2,2] - matA[1,2]*matA[2,1])
  b[1,3] <- numb13 / det1a
  b[2,3] <- numb23 / det1a
  b[3,3] <- numb33 / det2a
  print(b)
  return(b)

}
```

```r
M = runif(3, 2, 5)
N = matrix(M, 3, 3)
invert_matrix3x3(N)
solve(N)
chol2inv(N)


```

```r
# 2.  Essayer de mettre en évidence à quelle vitesse par rapport à n (est-ce en n, n2, n3, n2.3? ) l'algorithme est capable d'inverser la matrice (on pourra penser travailler en logarithme et faire une représentation graphique...).

a <- matrix(c(5, 1, 0, 3,-1, 2, 4, 0,-1), nrow=3, byrow=TRUE) 
require(MASS)
require(pracma)

all.equal(inv(a), ginv(a))
all.equal(ginv(a), solve(a))
all.equal(solve(a), invert_matrix3x3(a))

system.time(replicate(1000, ginv(a)))

system.time(replicate(1000, inv(a)))


system.time(replicate(1000, solve(a)))

system.time(replicate(1000, chol2inv(a)))

system.time(replicate(1000, invert_matrix3x3(a)))
```

```r
install.packages('microbenchmark')
library(microbenchmark)
boxplot(microbenchmark(invert_matrix3x3(a), inv(a), ginv(a), solve(a), chol2inv(a), times=10000L))
```
